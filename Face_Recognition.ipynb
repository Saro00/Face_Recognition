```python
import pafy
import cv2
import numpy as np
```

```python
url = 'https://www.youtube.com/watch?v=vyzUlMPEV2k'     #load video from YouTube
vPafy = pafy.new(url)
play = vPafy.getbest(preftype="mp4")
```

```python
print(vPafy.title)    #get information about video
print(vPafy.rating)
print(vPafy.viewcount)
print(vPafy.author)
print(vPafy.length)
print(vPafy.description)
```

```python
#get prediction of the video


cap = cv2.VideoCapture(play.url)  

cap.set(3, 480) #width of the frame
cap.set(4, 640) #height of the frame

MODEL_MEAN_VALUES = (78.4263377603, 87.7689143744, 114.895847746)
age_list = ['(0, 2)', '(4, 6)', '(8, 12)', '(15, 20)', '(25, 32)', '(38, 43)', '(48, 53)', '(60, 100)']
gender_list = ['Male', 'Female']
emo_list = [ 'Angry' , 'Disgust' , 'Fear' , 'Happy'  , 'Neutral' ,  'Sad' , 'Surprise']


def load_caffe_models():
    age_net = cv2.dnn.readNetFromCaffe('deploy_age.prototxt', 'age_net.caffemodel')
    gender_net = cv2.dnn.readNetFromCaffe('deploy_gender.prototxt', 'gender_net.caffemodel')
    emo_net = cv2.dnn.readNetFromCaffe('deploy_emo.prototxt', 'EmotiW_VGG_S.caffemodel')
    return(age_net, gender_net, emo_net)



age_net, gender_net, emo_net = load_caffe_models()
    
while not (cv2.waitKey(1) & 0xFF == ord('q')): 

    ret, image = cap.read()

    face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    cv2.imshow('frame', gray)
    faces = face_cascade.detectMultiScale(gray, 1.1, 5)

    if(len(faces)>0):
        print("Found {} faces".format(str(len(faces))))

    for (x, y, w, h ) in faces:
        cv2.rectangle(image, (x, y), (x+w, y+h), (255, 255, 0), 2)#Get Face 
        face_img = image[y:y+h, h:h+w].copy()
        blob = cv2.dnn.blobFromImage(face_img, 1, (227, 227), MODEL_MEAN_VALUES, swapRB=False)

    #Predict Gender
        gender_net.setInput(blob)
        gender_preds = gender_net.forward()
        gender = gender_list[gender_preds[0].argmax()]
        print("Gender : " + gender)

    #Predict Age
        age_net.setInput(blob)
        age_preds = age_net.forward()
        age = age_list[age_preds[0].argmax()]
        print("Age Range: " + age)
        
    #Predict Emotion
        emo_net.setInput(blob)
        emo_preds = emo_net.forward()
        emo = emo_list[emo_preds[0].argmax()]
        print("Emotion: " + emo)
        

        overlay_text = "%s %s %s" % (gender, age, emo)
        font = cv2.FONT_HERSHEY_SIMPLEX
        cv2.putText(image, overlay_text, (x, y), font, 1, (255, 255, 255), 2, cv2.LINE_AA)

        
    cv2.imshow('frame', image)
```
